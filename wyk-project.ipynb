```python
fig, ax = plt.subplots(1, 2)
 
ax[0].hist(red.alcohol, 10, facecolor ='red',
              alpha = 0.5, label ="Czerwone wine")
 
ax[1].hist(white.alcohol, 10, facecolor ='white',
           ec ="black", lw = 0.5, alpha = 0.5,
           label ="Biale wine")
 
fig.subplots_adjust(left = 0, right = 1, bottom = 0,
               top = 0.5, hspace = 0.05, wspace = 1)
 
ax[0].set_ylim([0, 1000])
ax[0].set_xlabel("Alcohol w % Proc")
ax[0].set_ylabel("Częstotliwość")
ax[1].set_ylim([0, 1000])
ax[1].set_xlabel("Alcohol w % Proc")
ax[1].set_ylabel("Crequency")
 
fig.suptitle("Dystrybucia Alcohol w  % Proc")
plt.show()
```

```python

red['type'] = 1
white['type'] = 0

wines = red.append(white, ignore_index = True)

from sklearn.model_selection import train_test_split
X = wines.iloc[:, 0:11]
y= wines['type']

X_train, X_test, y_train, y_test = train_test_split(
		X, y, test_size = 0.34, random_state = 45)
```

```python
#Strukturujem siec dla klasyfikacji 
```

```python

from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(12, activation ='relu', input_shape =(11, )))
model.add(Dense(9, activation ='relu'))
model.add(Dense(1, activation ='sigmoid'))
model.output_shape
model.summary()
model.get_config()

model.compile(loss ='binary_crossentropy',
optimizer ='adam', metrics =['accuracy'])
```

```python
model.fit(X_train, y_train,epochs=10, batch_size=1, verbose=1)
```

```python
#Przedwidziwamy typ win
```

```python
y_pred = model.predict(X_test)
#zaokrąglamy y_pred 
y_pred = [round(x[0]) for x in y_pred]
```

```python
score = model.evaluate(X_test, y_test,verbose=1)
```
